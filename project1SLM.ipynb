```python
import random
from collections import defaultdict
import re
from pprint import pprint
```

NOW WE NEED A FUNCTION TO READ THE TEXT FROM A FILE

```python
def load_text_from_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()
```

NOW WE MAKE A FUNCTION THAT PROCESSES THE GIVEN TEXT AND MAKES IT INTO A DICTIONARY

```python
def parse_text(text):
    # Split the input text into individual words
    words = text.split()

    # Create pairs of consecutive words (bigrams) except for the last two words to avoid index out of range
    word_pairs = [(words[i], words[i+1]) for i in range(len(words)-2)]

    # Initialize a defaultdict to hold lists of words that follow each bigram
    word_dict = defaultdict(list)

    # Iterate over the bigrams and the words that follow each bigram
    for pair, next_word in zip(word_pairs, words[2:]):
        # For each bigram, append the following word to the list in the dictionary
        word_dict[pair].append(next_word)

    # Return the populated dictionary
    return word_dict

# Example of how to use the function (commented out)
# source = 'message-2.txt'  # File path to the source text
# input_text = load_text_from_file(source)  # Load the content of the file
# parsed_text = parse_text(input_text)  # Parse the loaded text to find bigrams and their following words
# pprint(parsed_text)  # Pretty print the resulting dictionary
```

This function generates a sequence of words text based on a given dictionary of word pairs.

```python
import random

def predict_next_word(word_dict, starting_pair=None):
    # If no starting pair is provided, select one using select_starting_pair function
    if starting_pair is None:
        starting_pair = select_starting_pair(word_dict)

    # Initialize the output with the starting pair of words
    output = list(starting_pair)

    # Loop indefinitely to generate words based on the Markov chain
    while True:
        # Retrieve the list of possible next words for the current starting pair
        next_words = word_dict[starting_pair]

        # Randomly select one of the next possible words
        next_word = random.choice(next_words)

        # Append the selected next word to the output list
        output.append(next_word)

        # Check if there is a subsequent pair in the dictionary for the current pair and the next word
        # If yes, update the starting pair to this new pair (for the next iteration)
        if (starting_pair[1], next_word) in word_dict:
            starting_pair = (starting_pair[1], next_word)
        else:
            # If no subsequent pair is found, break the loop as the chain cannot continue
            break

    # Join the words in the output list into a single string separated by spaces and return it
    return ' '.join(output)
```

The select_starting_pair function is designed to select and return a starting pair of words from the word_dict

```python
def select_starting_pair(word_dict):
    # Extract keys that start with a capital letter
    capital_keys = [key for key in word_dict.keys() if key[0][0].isupper()]

    # Check if there are any capital keys
    if capital_keys:
        # If there are capital keys, randomly choose one
        return random.choice(capital_keys)
    else:
        # If there are no capital keys, choose randomly from all keys
        return random.choice(list(word_dict.keys()))
```


The provided function, predict_next_n_words, generates a sequence of words, given a dictionary mapping word pairs to lists of potential subsequent words. we used a AI in this function to help with some specific problem.

```python
def predict_next_n_words(word_dict, n, starting_pair=None):
    # Generate initial words based on starting pair (if provided) or randomly
    output = predict_next_word(word_dict, starting_pair).split()

    # Continue predicting until desired number of words (n) is reached
    while len(output) < n:
        # Check if the desired number of words is reached to prevent overshooting
        if len(output) >= n:
            break

        # Form a current pair of words for prediction
        current_pair = (output[-2], output[-1])

        # Randomly choose the next word based on the current pair
        next_word = random.choice(word_dict[current_pair])

        # Append the predicted word to the output sequence
        output.append(next_word)

    # Join the predicted words and return only the first n words
    return ' '.join(output[:n])
```

```python
def load_input_text(source):
    return load_text_from_file(source)

# Example usage
source = 'message-2.txt'  # Specify the input file path or URL
input_text = load_input_text(source)
parsed_text = parse_text(input_text)
pprint(parsed_text)

# Predict next word or N words
predicted_word = predict_next_word(parsed_text)
predicted_text = predict_next_n_words(parsed_text, 100)
print("Predicted Word:", predicted_word)
print("Predicted Text:", predicted_text)
```
